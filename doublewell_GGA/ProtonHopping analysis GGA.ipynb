```python
Energy_curves=[]
Pos=[]
for i in range(size(sep_mult)):
    energy_prime=[]
    for j in range(0,2*sep_mult[i]-1):
        with open(str('%.4f' % separations2[i])+"_"+str('%.4f' %positions2[i][j])+"_H2.Ecomponents") as f:
            lineList = f.readlines()
        energy_prime.append(float(lineList[-1].split()[-1]))
        #print(positions2[i][j])
    k=2*sep_mult[i]-1
    with open(str('%.4f' % separations2[i])+"_"+str('%.4f' %positions2[i][k])+"_H2.Ecomponents") as f:
            lineList = f.readlines()
            
    x0=float(float(lineList[-1].split()[-1]))
    Energy_curves.append(energy_prime+[x0]+energy_prime[::-1])
    Pos.append(positions2[i]+list(-array((positions2[i][:-1][::-1]))))
    #print(positions2[i][k],'\n')
```
